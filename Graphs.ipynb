## Categorical Data:
Categorical data refers to a form of information that can be stored and identified based on their names or labels. It is a type of qualitative data that can be grouped into categories instead of being measured numerically.
#### 1. Frequency Distribution
It is the number of times each value occurs in a dataset. We can use a barchart to represent this type of data. 


```python
plt.bar(data["Vacation"] , data["Count"],width=0.2 , color = ["Red", "Blue","Green", "Pink", "Yellow"])
```

In the above example, we are plotting a barchart for the dataset `data` imported using the **pandas** library. We can answer various questions like "What is the most visited destination/ the least visited destination" by looking at the barchart.

#### 2. Relative Frequency:
It is the proportion of a category in a dataset. It is calculated by dividing the frequency of a category by the total number observations in a dataset.

```python
plt.bar(data["Vacation"] , (data["Count"]/data["Count"].sum()),width=0.2 , color = ["Red", "Blue","Green", "Pink", "Yellow"])
```

In the above example, we are calculating the relative frequency by **data["Count"]/data["Count"].sum())**. WE mostly use Pie charts to represent this.

```python
plt.pie((data["Count"]/data["Count"].sum()), labels=data["Vacation"])
plt.title('Pie Chart - Vacation Dist')
```

#### 3. Cumalative Frequency
It is the running total of of a variable in a dataset. It is obtained by adding up the frequencies of the current category and all the frequency of all the previous categories.

```python
plt.bar(data["Vacation"] , data["Count"],width=0.2 , color = ["Red", "Blue","Green", "Pink", "Yellow"])
plt.plot(data["CumFreq"])
plt.title('Cumalative Freq - Vacation Dist')
plt.show()
```

We are plotting the Cumalative frequency using a bar and Ogive

## Numerical Data
#### Frequency Distribution Table and Histogram

```python
import numpy as np
```

```python
# Generate synthetic data
np.random.seed(42)
num_samples = 50
names = ['Person{}'.format(i) for i in range(num_samples)]
ages = np.random.randint(18, 70, num_samples)

# Create a DataFrame
data = {'Name': names, 'Age': ages}
df = pd.DataFrame(data)

# Plot histogram
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('No. of people')

plt.show()


```

The Histogram shows the distribution of people among various Age groups.

# Graphs for Bivariate Analysis:
### 1. Categorical-Categorical
A contigency table, also known as crosstab, is a type of table used in statistics to summerize the relationship between two categorical  variables.

```python
crossTab = pd.crosstab(Survive["Survived"],Survive["Pclass"])
```

```python
crossTab
```

```python
crossTab.plot(kind = "bar")
```

By obeserving the crosstab plot we can see the relationship between the Class of a Passenger and the survival chances of the passenges.
